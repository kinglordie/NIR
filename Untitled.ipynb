#      

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris ['target'], test_size=0.3, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

#  KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
y_pred=knn.predict(x_test)
```

```python
prediction_output=pd.DataFrame(data=[y_test.data, y_pred], index=['Actual', 'Predicted'])
prediction_output.transpose()
```

```python
from sklearn.metrics import accuracy_score
print ("accuracy: ", metrics.accuracy_score(y_test, y_pred))
print (metrics.classification_report(y_test, y_pred))
```

#   

```python
from sklearn.tree import DecisionTreeClassifier
dectree=DecisionTreeClassifier()
```

```python
dectree.fit(x_train,y_train)
y_pred1=dectree.predict(x_test)
```

```python
prediction_output1=pd.DataFrame(data=[y_test.data, y_pred1], index=['Actual', 'Predicted'])
prediction_output1.transpose()
```

```python
from sklearn.metrics import accuracy_score
print ("accuracy: ", metrics.accuracy_score(y_test, y_pred1))
print (metrics.classification_report(y_test, y_pred1))
```
